# La Liga (1995 - 2020)

# Import the libraries

```python
import pandas as pd
import numpy as np
```

# Import the dataset

```python
data = pd.read_csv('LaLiga_Matches_1995-2020.csv')
```

```python
data
```

# Goals scored in home matches in each season

```python
pivot1 = pd.pivot_table(data,
                       index=['Season'],
                       values=['FTHG'],
                       columns=['HomeTeam'],
                       fill_value='',
                       aggfunc=np.sum)
```

```python
pivot1
```

```python
pivot1['FTHG']['Barcelona'].plot(figsize=(15,8))
```

```python
pivot1 = pd.pivot_table(data,
                       index=['Season'],
                       values=['FTHG'],
                       columns=['HomeTeam'],
                       fill_value='',
                        aggfunc=np.sum, 
               margins=True, 
               margins_name='Total'
              )
```

```python
pivot1 
```

# Goals scored in away matches  in each season

```python
pivot3 = pd.pivot_table(data,
                       index=['Season'],
                       values=['FTAG'],
                       columns=['AwayTeam'],
                       fill_value='',
                        aggfunc=np.sum
              )
```

```python
pivot3
```

```python
pivot3['FTAG']['Barcelona'].plot(figsize=(15,8))
```

```python
pivot4 = pd.pivot_table(data,
                       index=['Season'],
                       values=['FTAG'],
                       columns=['AwayTeam'],
                       fill_value='',
                        aggfunc=np.sum, 
               margins=True, 
               margins_name='Total'
              )
```

```python
pivot4
```
