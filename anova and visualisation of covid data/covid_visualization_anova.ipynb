# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Datasets

```python
dataset = pd.read_csv('us_counties_covid19_daily.csv')
dataset
```

```python
df = dataset[:]
```

# To know the missing values

```python
df.isna().sum()
```

```python
df['deaths'].fillna(0,inplace=True)
df
```

```python
#To know the missing values
df.isna().sum()
```

```python
df.state.unique()
```

```python
df2 = df[df['date']=='2020-12-05']
df2
```

```python
df3=df2.groupby('state')['cases','deaths'].sum().reset_index()
df3
```

```python
fig = plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
plt.bar(df3['state'],df3['cases'])
plt.ylabel('Cases')
plt.title('State vs Cases (Bar chart)')
plt.show()
```

```python
fig = plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
plt.bar(df3['state'],df3['deaths'])
plt.ylabel('Deaths')
plt.title('State vs Deaths (Bar chart)')
plt.show()
```

```python
df4=df3.copy()
df4['Mortality_rate']=df4['deaths']/df4['cases']
df4
```

```python
fig = plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
plt.bar(df4['state'],df4['Mortality_rate'],edgecolor='k')
plt.ylabel('Mortality Rate')
plt.title('State vs Mortality rate (Bar chart)')
plt.show()
```

```python
fig = plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
plt.plot(df4['state'],df4['Mortality_rate'])
plt.ylabel('Mortality Rate')
plt.title('State vs Mortality rate (Line Plot)')
plt.show()
```

# ANOVA

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
mod=ols('cases~Mortality_rate',data=df4).fit()
aov=sm.stats.anova_lm(mod,type=2)
print(aov)
```
